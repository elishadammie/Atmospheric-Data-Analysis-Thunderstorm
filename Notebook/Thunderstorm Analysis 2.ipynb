```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_graph(w, x, y, z, name):
    
    w_ = df.loc[w]
    x_ = df.loc[x]
    y_ = df.loc[y]
    z_ = df.loc[z]
    
    fig = plt.figure(figsize=(15,8))
    fig.suptitle('Thunderstom strike against Month for {}, {}, {} and {} in {}'.format(str(w), str(x), str(y), str(z), str(name)[:-5]), fontsize=16)
    plt.ylabel('Thunderstom Strike', fontsize=12)
    plt.xlabel('Months of the Year', fontsize=12)

    plt.plot(w_, )
    plt.plot(x_, )
    plt.plot(y_, )
    plt.plot(z_, )
    plt.legend([str(w), str(x), str(y), str(z)])
    
    plt.savefig('{}.svg'.format(str(name)[:-5]), dpi=300)
```

```python
path = os.getcwd()
files = os.listdir(path)
files

files_xls = [f for f in files if f[-4:] == 'xlsx']


for item in files_xls:
    df = pd.read_excel(str(item))

    df.rename_axis(['Year', ])

    newcol = 'YEAR JAN FEB MAR APR MAY JUN JUL AUG SEP OCT NOV DEC'.split(' ')

    df.set_axis(newcol, axis=1, inplace=True)

    df.set_index(['YEAR'], inplace=True)
    
    plot_graph(2015, 2016, 2017, 2018, item)
```
