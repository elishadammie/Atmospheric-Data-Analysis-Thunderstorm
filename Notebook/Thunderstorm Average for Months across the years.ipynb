```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pprint
```

```python
path = os.getcwd()
files = os.listdir(path)
files

files_xls = [f for f in files if f[-4:] == 'xlsx']

for item in files_xls:
    df = pd.read_excel(str(item))
    df.rename_axis(['Year', ])

    newcol = 'YEAR JAN FEB MAR APR MAY JUN JUL AUG SEP OCT NOV DEC'.split(' ')
    df.set_axis(newcol, axis=1, inplace=True)
    df.set_index(['YEAR'], inplace=True)
    
    labels = newcol[1:]

    fig = plt.figure(figsize=(15,8))
    fig.suptitle('Average Monthly Thunderstorm Frequency for {} over the years'.format(str(item)[:-5]), fontsize=16)
    plt.ylabel('Thunderstorm Strike', fontsize=12)
    plt.xlabel('Months of the Year', fontsize=12)
    
    
    bp_dict = plt.bar(labels, list(df.mean(axis=0)), align='edge', width=-0.4)
```
